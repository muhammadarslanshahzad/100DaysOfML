```python
from sklearn.datasets import load_boston
```

```python
X , y = load_boston(return_X_y=True) 
```

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pylab as plt
```

```python
mod=KNeighborsRegressor().fit(X, y)
```

```python
pred = mod.predict(X)
```

```python
plt.scatter(pred, y)
```
